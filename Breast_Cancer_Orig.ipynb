```python
#Need to recode data to be numeric rather than string
#Start with binary variables, recode to 0/1
def binary_recode(boolean):
    if boolean == "Yes":
        return 1
    else:
        return 0
def boolean_recode(boolean):
    if boolean == True:

        return 1
    if boolean == False:
        return 0
    else:
        print("ERROR in value, ", boolean)
        return -1
    
#Source - #https://www.cancer.org/cancer/breast-cancer/screening-tests-and-early-detection/mammograms/understanding-your-mammogram-report.html

def birad_assess(result):
    if result == "Needs additional imaging":
        return 0
    if result == "Negative":
        return 1
    if result == "Benign findings":
        return 2
    if result == "Probably benign":
        return 3
    if result == "Suspicious abnormality":
        return 4
    if result == "Highly suggestive of malignancy":
        return 5
    if result == "Known biopsy-proven malignancy":
        return 6
    else:
        print("ERROR in value ", result)
        return 10

#Source -- https://www.ncbi.nlm.nih.gov/books/NBK343794/table/ch1.t1/
def birad_density(density):
    if density == "Almost entirely fatty":
        return 1
    if density == "Scattered fibroglandular densities":
        return 2
    if density == "Heterogeneously dense":
        return 3
    if density == "Extremely dense":
        return 4
    else:
        print("ERROR in value ", density)
        return 10
def cancer_type(cancer):
    if cancer == "No cancer diagnosis":
        return 1
    if cancer == "Invasive cancer":
        return 2
    if cancer == "ductal carcinoma in situ":
        return 3
    else:
        print(cancer)
        return -1

    
data['History_Of_Breast_Biopsy_Clean'] = data['History_Of_Breast_Biopsy'].apply(binary_recode)
data['Comparison_Mammogram_From_Mammography_Clean'] = data['Comparison_Mammogram_From_Mammography'].apply(binary_recode)
data['Binary_Indicator_Clean'] = data['Binary_Indicator'].apply(binary_recode)
data['Family_History_Of_Breast_Cancer_Clean'] = data['Family_History_Of_Breast_Cancer'].apply(binary_recode)
data['Current_Use_Of_Hormone_Therapy_Clean'] = data['Current_Use_Of_Hormone_Therapy'].apply(binary_recode)
#print("1")
data['Is_Film_Or_Digital_Mammogram_Clean'] = data['Is_Film_Or_Digital_Mammogram'].apply(boolean_recode)
data['Is_Binary_Indicator_Of_Cancer_Diagnosis_Clean'] = data['Is_Binary_Indicator_Of_Cancer_Diagnosis'].apply(boolean_recode)
data['Radiologists_Assessment_Clean'] = data['Radiologists_Assessment'].apply(birad_assess)
data['Patients_BI_RADS_Breast_Density_Clean'] = data['Patients_BI_RADS_Breast_Density'].apply(birad_density)
data['Cancer_Type_Clean'] = data['Cancer_Type'].apply(cancer_type)

import seaborn as sns
#sns.histplot(x=data['Is_Film_Or_Digital_Mammogram'])

#data.head()
```

```python
#Now we create a new dataframe with the clean data
#We Drop BMI due to missing data
clean_data = data.drop(columns = ['Radiologists_Assessment', 'Body_Mass_Index', 'Cancer_Type' ,'Is_Binary_Indicator_Of_Cancer_Diagnosis', 'Comparison_Mammogram_From_Mammography', 'Patients_BI_RADS_Breast_Density','Family_History_Of_Breast_Cancer', 'Current_Use_Of_Hormone_Therapy', 'Binary_Indicator','History_Of_Breast_Biopsy','Is_Film_Or_Digital_Mammogram'])
clean_data.head()
#clean_data.shape
```

```python
#Logistic Regression Model
X = clean_data.drop(columns='Is_Binary_Indicator_Of_Cancer_Diagnosis_Clean')
y = clean_data['Is_Binary_Indicator_Of_Cancer_Diagnosis_Clean']
```

```python
#Using 20 percent for model testing and 80% for training
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
from sklearn.linear_model import LogisticRegression
model_log = LogisticRegression()
model_log.fit(X_train,y_train)
y_pred = model_log.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
```
